### Import the data

```python
import pandas as pd
import numpy as np
concrete_data = pd.read_csv('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DL0101EN/labs/data/concrete_data.csv')
concrete_data.head()
```

### Dataframe Split

In this step, I split the Dataframe into 2 parts: 

1. dataframe of predictions;

2. vector of target.

```python
concrete_data_columns = concrete_data.columns

predictors = concrete_data[concrete_data_columns[concrete_data_columns != 'Strength']] 
target = concrete_data['Strength'] 
```

Now, I show you the head of 2 parts:

```python
predictors.head()
```

```python
target.head()
```

I calculate the numbers of columns that I will use in the next steps to create a neural network.

```python
n_cols = predictors.shape[1]
n_cols
```

## Part A

### Import Keras libraries

```python
import keras
from keras.models import Sequential #the network is a linear stack of layers
from keras.layers import Dense #for create layers
```

The are two way to create a `Sequential( )` model:

1. by passing a list of layer instances to the constructor;

2. add layer via the `add()` method.

(Look at link: https://keras.io/getting-started/sequential-model-guide/)

I choose the second one way.

The **neural network** should be have:

- One hidden layer of 10 nodes;
- a ReLU activation function;
- adam algorithm like optimizer (it not specify the learning rate);
- mean squared error (error between predictive value and truth value) as the loss function.

#### Create the model

```python
#Create the model with one hidden layer
model = Sequential()

model.add(Dense(10, activation = 'relu', input_shape = (n_cols,))) #create hidden layer
model.add(Dense(1)) #create output layer
```

#### Compile the model

```python
model.compile(optimizer = 'adam', loss = 'mean_squared_error')
```

```python
print(model.summary())
```

### 1A. Randomly split the data into a training and test sets by holding 30% of the data for testing

I use the `train_test_split( )` function (from Scikit-learn) to generate the random split of the data into a training and test sets by holding 30% of the data for testing.

```python
import numpy as np
from sklearn.model_selection import train_test_split
xTrain, xTest, yTrain, yTest = train_test_split(predictors, target, test_size = 0.3, random_state = 0)
```

Here I plot the shape of the splitting data:

```python
xTrain.shape, yTrain.shape
```

```python
xTest.shape, yTest.shape
```

### 2A. Train the model on the training data using 50 epochs

```python
# fit the model: leave out 30% of the data for validation and we will train the model for 50 epochs
model.fit(xTrain, yTrain, epochs=50)
```

### 3A. Evaluate the model on the test data and compute the mean squared error between the predicted concrete strength and the actual concrete strength

I predict the value using the test data.

```python
#Predict Strength
y_pred = model.predict(xTest)
```

And I valuate the model with the `mean_squared_error( )` function from sklearn.metrics.

```python
# Evaluate the model with mean squared error
from sklearn.metrics import mean_squared_error
MSE = mean_squared_error(yTest, y_pred)
```

Here, the results:

```python
print('The Mean squered error is:', MSE)
```

### 4A. Repeat steps 1 - 3, 50 times, i.e., create a list of 50 mean squared errors

```python
#create a MSE array
MSE_50_A = []
for i in range(50):
    xTrain, xTest, yTrain, yTest = train_test_split(predictors, target, test_size = 0.3, random_state = 0)
    model.fit(xTrain, yTrain, epochs=50)
    y_pred = model.predict(xTest)
    MSE_50_A.append(mean_squared_error(yTest, y_pred))
```

I check the length of MSE, with the function, `len( )`:

```python
len(MSE_50_A)
```

### 5A. Report the mean and the standard deviation of the mean squared errors

```python
import statistics 
```

```python
# Calculate the Mean od MSE_50
MEAN_MSE = statistics.mean(MSE_50_A)
```

```python
# Calculate the Standard Deviation of MSE_50
STD_DEV = statistics.stdev(MSE_50_A)
```

```python
print('The mean of MSE_A is:',MEAN_MSE,'.','The Standard Deviation of MSE_A is',STD_DEV)
```

## Part B

Using the same neural network of part A, I test the model with normalized values of predictors and targets. Then I calculate the average value of MSE and Standard Deviation.

### Normalize the Data

#### Predictors data

```python
#Predictors
predictors_norm = (predictors - predictors.mean()) / predictors.std()
predictors_norm.head()
```

#### Targets data

```python
#targets
targets_norm = (target - target.mean()) / target.std()
targets_norm.head()
```

### 1B. Randomly split the data into a training and test sets by holding 30% of the data for testing

```python
#random split
import numpy as np
from sklearn.model_selection import train_test_split
xTrain_norm, xTest_norm, yTrain_norm, yTest_norm = train_test_split(predictors_norm, targets_norm, test_size = 0.3, random_state = 0)
```

### 2B. Train the model on the training data using 50 epochs

```python
# fit the model: leave out 30% of the data for validation and we will train the model for 50 epochs
model.fit(xTrain_norm, yTrain_norm, epochs=50)
```

### 3B. Evaluate the model on the test data and compute the mean squared error between the predicted concrete strength and the actual concrete strength.

```python
#Predict Strength
y_pred_norm = model.predict(xTest_norm)
```

```python
# Evaluate the model with mean squared error
from sklearn.metrics import mean_squared_error
MSE_norm = mean_squared_error(yTest_norm, y_pred_norm)
```

### 4B. Repeat steps 1 - 3, 50 times, i.e., create a list of 50 mean squared errors

```python
#create a MSE array
MSE_50_norm_B = []
for i in range(50):
    xTrain_norm, xTest_norm, yTrain_norm, yTest_norm = train_test_split(predictors_norm, targets_norm, test_size = 0.3, random_state = 0)
    model.fit(xTrain_norm, yTrain_norm, epochs=50)
    y_pred_norm = model.predict(xTest_norm)
    MSE_50_norm_B.append(mean_squared_error(yTest_norm, y_pred_norm))
    
```

### 5B. Report the mean and the standard deviation of the mean squared errors.

```python
import statistics 
```

```python
# Calculate the Mean od MSE_50
MEAN_MSE_norm_B = statistics.mean(MSE_50_norm_B)
# Calculate the Standard Deviation of MSE_50
STD_DEV_norm_B = statistics.stdev(MSE_50_norm_B)
print('The mean of MSE_B is:',MEAN_MSE_norm_B,'.','The Standard Deviation of MSE_B is',STD_DEV_norm_B)
```

## How does the mean of the mean squared errors compare to that from Step A?

## Part D

I reconstructed the neural network with 3 hidden layers and tested the model with the values of part B. Also here I repeated only points 4 and 5.

### Import Keras libraries and creation of new neural network

```python
#import keras libraries
import keras
from keras.models import Sequential #the network is a linear stack of layers
from keras.layers import Dense #for create layers
```

The **neural network** should be have:

- 3 hidden layer of 10 nodes;
- a ReLU activation function;
- adam algorithm like optimizer (it not specify the learning rate);
- mean squared error (error between predictive value and truth value) as the loss function.

```python
#create model with one hidden layer
modelD = Sequential()
modelD.add(Dense(10, activation='relu', input_shape=(n_cols,)))
modelD.add(Dense(10, activation='relu', input_shape=(n_cols,)))
modelD.add(Dense(10, activation='relu', input_shape=(n_cols,)))
modelD.add(Dense(1))
```

```python
print(modelD.summary())
```

```python
modelD.compile(optimizer = 'adam', loss = 'mean_squared_error')
```

### 4D. Repeat steps 1 - 3, 50 times, i.e., create a list of 50 mean squared errors (with modelD)

```python
#create a MSE array
MSE_50_norm_D = []
for i in range(50):
    xTrain_norm, xTest_norm, yTrain_norm, yTest_norm = train_test_split(predictors_norm, targets_norm, test_size = 0.3, random_state = 0)
    modelD.fit(xTrain_norm, yTrain_norm, epochs=50)
    y_pred_normD = modelD.predict(xTest_norm)
    MSE_50_norm_D.append(mean_squared_error(yTest_norm, y_pred_norm))
```

### 5C. Report the mean and the standard deviation of the mean squared errors.

```python
import statistics 
# Calculate the Mean od MSE_50
MEAN_MSE_norm_D = statistics.mean(MSE_50_norm_D)
# Calculate the Standard Deviation of MSE_50
STD_DEV_norm_D = statistics.stdev(MSE_50_norm_D)
print('The mean of MSE_C is:',MEAN_MSE_norm_D,'.','The Standard Deviation of MSE_C is',STD_DEV_norm_D)
```

## How does the mean of the mean squared errors compare to that from Step B?

```python
print('The mean of MSE_B is:',MEAN_MSE_norm_B)
print('The mean of MSE_D is;',MEAN_MSE_norm_D)
```
